```python
import numpy as np
import pandas as pd
import lime
import lime.lime_tabular

import sklearn
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

# Fetching data and visualizing data

This data is availabe at: https://www.kaggle.com/uciml/breast-cancer-wisconsin-data

```python
cancer_data = pd.read_csv('../datasets/breast_cancer.csv')
```

### Create 3 dataframes for each type of feature: mean, se and worst

```python
cancer_data.mean = cancer_data.filter(regex='_mean$')
cancer_data.se = cancer_data.filter(regex='_se$')
cancer_data.worst = cancer_data.filter(regex='_worst$')
```

### Model training

The first mode will be trained with the mean dataset using all variables.

```python
random_seed = 13

y = cancer_data['diagnosis'].values

X_train, X_test, y_train, y_test = train_test_split(cancer_data.mean, y, test_size=0.25, random_state=random_seed)
```

```python
rf = RandomForestClassifier(n_estimators=20, random_state=random_seed)
rf.fit(X_train, y_train)
```

```python
metrics.accuracy_score(y_test, rf.predict(X_test))
```

### Create the explainer

For tabular data, LIME needs the dataset used for the training step.

```python
explainer = lime.lime_tabular.LimeTabularExplainer(X_train.values, list(X_train), class_names = ['M','B'], discretize_continuous=True)
explainer
```

### Explaining an instance

```python
idx = 13
exp = explainer.explain_instance(X_test.values[idx], rf.predict_proba, num_features=4)
```
